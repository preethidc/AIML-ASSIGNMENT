```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "Alex drives to an awful job every morning",
    "Very few giant animals remain in the wild",
    "The little fluffy puppy is ready to adopt",
    "Jesica lives in a house that produces electricity",
    "The fast black fox jumped over the lazy dog"
]

vectorizer = TfidfVectorizer(stop_words='english')

X = vectorizer.fit_transform(documents)

feature_names = vectorizer.get_feature_names_out()

for doc_num, doc in enumerate(X.toarray()):

    print(f"\nDocument {doc_num+1}")

    scores = zip(feature_names, doc)

    sorted_scores = sorted(scores,
                           key=lambda x: x[1],
                           reverse=True)

    for word, score in sorted_scores[:5]:
        print(word, round(score, 3))
```
