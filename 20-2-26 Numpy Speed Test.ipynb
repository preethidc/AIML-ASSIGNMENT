```python
#Numpy speed test
import time
import numpy as np

# Create data
n = 1_000_000

# Python list
py_list = list(range(n))

# NumPy array
np_array = np.arange(n)

# --- Operation: multiply each element by 2 ---

# Python list timing
start = time.time()
result_list = [x * 2 for x in py_list]
end = time.time()
print("Python list time:", end - start)

# NumPy array timing
start = time.time()
result_array = np_array * 2
end = time.time()
print("NumPy array time:", end - start)


#1.NumPy arrays are much faster than Python lists for numerical operations
#2.NumPy arrays use less memory since all elements are of the same type, 
#while Python lists store extra information for each element.
#3.NumPy allows direct operations on entire arrays (np_array * 2),
#while Python lists require loops or list comprehensions.
```
