```python
from sklearn.tree import DecisionTreeClassifier

import matplotlib.pyplot as plt

X = [
    [0, 0, 0],  
    [0, 0, 1], 
    [0, 1, 0],  
    [1, 0, 0],  
    [2, 1, 0],  
    [2, 1, 1],  
    [2, 0, 0],  
]

# Output: Play (1) or Not Play (0)
y = [0, 0, 1, 1, 1, 0, 1]

model = DecisionTreeClassifier()

model.fit(X, y)

prediction = model.predict([[0, 1, 0]])

if prediction[0] == 1:
    print("Play Outside ")
else:
    print("Don't Play ")

```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plot_tree(model, feature_names=["Weather", "Humidity", "Wind"], class_names=["No", "Yes"], filled=True)
plt.show()
```
