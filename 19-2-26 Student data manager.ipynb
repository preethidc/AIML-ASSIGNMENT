```python
#Student Data Manager
students = {
    "Asha": 87,
    "Bindu": 98,
    "Chirag": 76,
    "Darshan": 85,
    "Elena": 99
}

# Topper
topper = max(students, key=students.get)
print("Topper:", topper, students[topper])

# Average
avg = sum(students.values()) / len(students)
print("Average:", avg)

# Grades
for name in students:
    marks = students[name]
    if marks >= 90:
        print(name, "Grade: A")
    elif marks >= 75:
        print(name, "Grade: B")
    else:
        print(name, "Grade: C")
```
