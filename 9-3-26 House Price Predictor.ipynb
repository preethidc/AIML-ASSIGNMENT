```python
import numpy as np
from sklearn.linear_model import LinearRegression
# Input-house size 
sizes = np.array([500, 800, 1000, 1200, 1500, 1800]).reshape(-1,1)

# Output-price
prices = np.array([50, 80, 100, 120, 150, 180])
model = LinearRegression()
model.fit(sizes, prices)

# Predict price for house
predicted_price = model.predict([[1300]])

print("Predicted Price:", predicted_price[0])

test_sizes = [[700], [1600], [2000]]
predictions = model.predict(test_sizes)

print("Predictions:", predictions)
```
