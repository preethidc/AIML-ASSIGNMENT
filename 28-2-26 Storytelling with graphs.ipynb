```python
#Storytelling with Graphs
import pandas as pd

data = {
    "Product": ["Shoes","Bags","Watches","Shoes","Bags","Watches","Shoes","Bags","Watches"],
    "Month": ["Jan","Jan","Jan","Feb","Feb","Feb","Mar","Mar","Mar"],
    "Units_Sold": [45,35,20,55,40,25,65,50,30],
    "Price": [1200,900,2000,1200,900,2000,1200,900,2000]
}

df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
```

```python
product_sales = df.groupby("Product")["Units_Sold"].sum()

product_sales.plot(kind="bar")
plt.title("Total Units Sold by Product")
plt.xlabel("Product")
plt.ylabel("Units Sold")
plt.show()
```

```python
product_sales.plot(kind="pie", autopct="%1.1f%%")
plt.title("Units Sold Distribution by Product")
plt.ylabel("")
plt.show()
```

```python
plt.hist(df["Units_Sold"])
plt.title("Distribution of Units Sold")
plt.xlabel("Units Sold")
plt.ylabel("Frequency")
plt.show()
```

```python
#The dataset shows sales trends for Shoes, Bags, and Watches over three months. 
#The bar chart indicates that Shoes are the most sold product, followed by Bags, while Watches have the lowest sales.
#The pie chart confirms that Shoes contribute the largest share of total units sold, showing strong customer demand. 
#Bags have moderate performance, and Watches lag behind in quantity sold.
#The histogram shows that most sales values fall between 30 and 60 units, indicating consistent sales across products and months.
#Additionally, sales for all products show a steady increase from January to March, suggesting growing demand over time.
#Overall, Shoes dominate in volume, while Bags show stable growth and Watches may need better marketing to improve sales
```
