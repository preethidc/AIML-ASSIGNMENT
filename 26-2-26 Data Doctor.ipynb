```python
#Data Doctor

import pandas as pd

df = pd.read_csv("students.csv")
print(df.isnull().sum())

df.fillna(0, inplace=True) 

df.drop_duplicates(inplace=True)
df["Name"] = df["Name"].str.lower()

# Remove extra spaces
df["Name"] = df["Name"].str.strip()

print(df.head())
df.dropna(inplace=True)
```
