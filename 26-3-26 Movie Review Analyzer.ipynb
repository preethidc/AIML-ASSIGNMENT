```python
# Import libraries
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Dataset
reviews = [
    "The movie was amazing and very entertaining",
    "I hated the film, it was boring and too long",
    "Fantastic acting and wonderful story",
    "The movie was terrible and disappointing",
    "An excellent and inspiring film"
]

sentiments = [
    "Positive",
    "Negative",
    "Positive",
    "Negative",
    "Positive"
]

# Convert text into numerical data
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(reviews)

# Train model
model = MultinomialNB()
model.fit(X, sentiments)

# Test reviews
test_reviews = [
    "The movie was fantastic",
    "I dislike this boring movie"
]

# Convert test reviews
test_X = vectorizer.transform(test_reviews)

# Predict sentiment
predictions = model.predict(test_X)

# Display results
for review, sentiment in zip(test_reviews, predictions):
    print("Review:", review)
    print("Predicted Sentiment:", sentiment)
    print()
```
