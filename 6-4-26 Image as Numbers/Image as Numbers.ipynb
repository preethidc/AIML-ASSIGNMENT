```python
import cv2
import numpy as np

# Load image
image = cv2.imread("sample.jpg")


# Print image shape
print("Image Shape:", image.shape)

# Print image dimensions
height, width, channels = image.shape
print("Height:", height)
print("Width:", width)
print("Channels:", channels)

# Print some pixel values
print("\nPixel at (0,0):", image[0,0])
print("Pixel at (50,50):", image[50,50])

# Print first 5x5 pixel values
print("\nFirst 5x5 Pixel Values:")
print(image[0:5, 0:5])

# Separate channels
blue = image[:, :, 0]
green = image[:, :, 1]
red = image[:, :, 2]

print("\nBlue Channel:")
print(blue)

print("\nGreen Channel:")
print(green)

print("\nRed Channel:")
print(red)
```
