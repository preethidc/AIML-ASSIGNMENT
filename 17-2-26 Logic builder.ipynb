```python
#17-2-26
#Logic Builder
def fizz_buzz():
    fizz = buzz = fizzbuzz = 0 #fizz divisible by 3,buzz divisible by 5,fizzbuzz divisible by both 3 and 5,,#initially set to 0
    
    for i in range(1, 51):    
        if i % 3 == 0 and i % 5 == 0: #divisibility check
            print("FizzBuzz")
            fizzbuzz += 1
        elif i % 3 == 0:
            print("Fizz")
            fizz += 1
        elif i % 5 == 0:
            print("Buzz")
            buzz += 1
        else:
            print(i)
            
    print("\nCounts:")
    print("Fizz:", fizz)
    print("Buzz:", buzz)
    print("FizzBuzz:", fizzbuzz)

fizz_buzz()
```
