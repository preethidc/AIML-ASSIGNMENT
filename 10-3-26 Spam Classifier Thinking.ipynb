```python
'''
SPAM CLASSIFIER THINKING

1.Problem Statement
Classify a message as:
Spam (unwanted/junk)
Ham (normal/legitimate)

2.Data Required
You need a labeled dataset like:
Message
Label
"Win ₹10,000 now!!! Click here"
Spam
"Hey, are we meeting today?"
Ham
"Limited offer!!! Buy now!!!"
Spam
Sources:
SMS datasets
Email datasets
Custom collected data

3.Data Preprocessing
Clean the text before using it:
Convert to lowercase
Remove punctuation
Remove stopwords (like "is", "the", "and")
Tokenization (split into words)
Stemming/Lemmatization

4.Feature Extraction
Convert text into numbers using:
Bag of Words (BoW)
TF-IDF (Term Frequency–Inverse Document Frequency)
Example:
"win money now" → [0,1,1,0,1,...]

5.Model Selection
Common algorithms:
Naive Bayes (best for text)
Logistic Regression
Decision Tree
Support Vector Machine (SVM)
Most commonly used: Naive Bayes

6.Model Training
Split dataset into:
Training set (80%)
Testing set (20%)
Train model on training data

7.Evaluation Metrics
Check model performance using:
Accuracy
Precision
Recall
F1 Score
Example:
High precision → fewer false spam alerts
High recall → catches most spam

8.Possible Mistakes
False Positive: Normal message marked as spam
False Negative: Spam message not detected
Example:
"Get free notes" → may be wrongly marked as spam

'''
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Data
messages = ["Win money now", "Hello friend", "Free offer now", "Let's study"]
labels = [1, 0, 1, 0]  # 1 = Spam, 0 = Ham

# Convert text to features
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(messages)

# Model
model = MultinomialNB()
model.fit(X, labels)

# Prediction
test = vectorizer.transform(["Win free money"])
print(model.predict(test))  # Output: [1] (Spam)
```
